```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Sleep_efficiency.csv")
df
```

```python
plt.scatter(df['Age'], df['Sleep duration'])
plt.xlabel('Age')
plt.ylabel('Sleep duration')
plt.title('Sleep duration of People by their Ages')
plt.show()
```

```python

plt.bar(df['Age'], df['Deep sleep percentage'])
plt.xlabel('Age')
plt.ylabel('Deep sleep percentage')
plt.title('Deep Sleep duration of People by their Ages')
plt.show()
```

```python
plt.bar(df['Gender'], df['Exercise frequency'])
plt.xlabel('Gender')
plt.ylabel('Exercise frequency')
plt.title('Which Gender Exercise frequently')
plt.show()
```
